```python
import pandas as pd
df = pd.read_excel('금천미트 구매후기.xlsx')
```

```python
df['부위'] = df['상품'].str.split('/').str[2]
df = df.drop('Unnamed: 0', axis =1)
```

```python
cleansing_data = df.drop_duplicates(subset=  ['날짜','부위', '작성자', '만족도'], inplace = False)
cleansing_data
```
